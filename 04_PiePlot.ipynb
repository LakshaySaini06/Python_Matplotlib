# Pie Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]
c = ["c","g","r","b"]
ex = [0.5 , 0 , 0.0 , 0.0]
plt.pie(x, labels = y, explode = ex , colors = c, autopct = "%0.1f%%", shadow = True, radius = 1.2, labeldistance = 1.1, startangle = 90
       ,textprops = {"fontsize":15, "color" :"y"}, counterclock = False
       ,wedgeprops = { "linewidth":5,"edgecolor": "m"}
       ,center = (1,3), rotatelabels = True)
#plt.title("Lakshay")
plt.legend(loc =1)
plt.show()
```
