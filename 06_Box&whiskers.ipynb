# Box and Whiskers Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70,200]
#vert = False
plt.boxplot(x, notch = True, widths = 0.5, label = ["Lakshay"], patch_artist = True, showmeans = True, whis = 1,sym = "g+", boxprops =dict(color = "r"),capprops = dict(color = "g"),
            whiskerprops =dict(color = "r"),flierprops=dict(markeredgecolor="y"))
plt.legend()
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]
x1 = [20,30,40,50,60,70 ,60]

y = [x,x1]
plt.boxplot(y, tick_labels= ["python","c++"])
plt.show()
```
