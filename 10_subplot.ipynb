# subplot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

x1 = [10,20,30,40]
y1 = ["a", "s", "d","f"]

plt.subplot(2,2,1)
plt.pie([1])


plt.subplot(2,2,2)
plt.bar(x1,y1)

plt.subplot(2,2,3)
plt.plot(x,y,color = "r")

plt.subplot(2,2,4)
plt.bar(x1,y1)

plt.show()
```
