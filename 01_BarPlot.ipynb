# BarPLot

```python
import matplotlib.pyplot as plt
```

```python
x = ["Python","C","C++","Java"]
y = [90,70,60,85]
z = [20,40,50,80]
plt.xlabel("Languages",fontsize = 10)
plt.ylabel("Popularity")
plt.title("Coding", fontsize = 20)
c = ["r","y","g","b"]
plt.bar(x,y,width = 0.4,color = c, align = "edge", edgecolor = "c", linewidth = 5, linestyle = ":", alpha = 0.7,label = "Lakshay") # by default center is used
plt.bar(x,z,width = 0.4,color = "b", align = "edge", edgecolor = "c", linewidth = 5, alpha = 0.7,label = "Saini")
#plt.bar(x,y,width = 0.4,color = c, align = "center")
#plt.bar(x,y,width = 0.4,color= "y")For single color
plt.legend()
plt.show()
```

```python
import numpy as np
x = ["Python","C","C++","Java"]
y = [90,70,60,85]
z = [20,40,50,80]
p = np.arange(len(x))
plt.xlabel("Languages",fontsize = 10)
plt.ylabel("Popularity")
plt.title("Coding", fontsize = 20)
c = ["r","y","g","c"]
width = 0.2
p1 = [j+width for j in p]
plt.bar(p,y,width,color = c, align = "edge", edgecolor = "c", linewidth = 5, linestyle = ":", alpha = 0.7,label = "Lakshay")
plt.bar(p1,z,width,color = "b", align = "edge", edgecolor = "c", linewidth = 5, alpha = 0.7,label = "Saini")
plt.xticks(p+width/2 , x,rotation = 20)
plt.legend()
plt.show()
```

```python
#horizontal bar graph
x = ["Python","C","C++","Java"]
y = [90,70,60,85]
z = [20,40,50,80]
c = ["r","y","g","c"]
plt.barh(x,y,color = c)
plt.barh(x,z,color ="c")
plt.show()
```
