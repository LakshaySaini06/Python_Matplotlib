# Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
area1 = [9,8,3,5,4]
area2 = [8,7,4,3,2]
area3 = [1,2,3,4,6]

l = ["area1","area2","area3"]
plt.stackplot(x,area1, area2, area3, labels = l,colors = ["r","g","b"], baseline = "wiggle")
plt.title("Lakshay")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid()
plt.legend(loc =1)
plt.show()
```
