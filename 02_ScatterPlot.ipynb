# ScatterPlot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,5,3,5,7,3,5]
y = [1,4,6,5,3,6,8]
plt.scatter(x,y,color = "r")
plt.title("Lakshay",fontsize = 20)
plt.xlabel("Day", fontsize = 10)
plt.ylabel("Hour",fontsize=10)
plt.show()
```

```python
x = [1,5,3,5,7,3,5]
y = [1,4,6,5,3,8,8]
y2= [3,2,5,6,7,8,9]
colors = [10,20,30,40,50,60,70]
sizes = [100,900,800,700,600,500,400]
plt.scatter(x,y,c=colors,s =sizes, alpha = 0.7,marker = "*",edgecolor = "g", linewidth = 3,cmap = "BuGn")
plt.scatter(x,y2)

t = plt.colorbar()
t.set_label("Saini",fontsize = 10)
plt.title("Lakshay",fontsize = 20)
plt.xlabel("Day", fontsize = 10)
plt.ylabel("Hour",fontsize=10)
plt.show()
```
