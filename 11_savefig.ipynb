# savefig

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,9,4,6,8]

plt.plot(x,y,color = "r")

plt.savefig("line")
plt.savefig("line1",dpi = 2000)
plt.savefig("line2",facecolor = "g")
plt.savefig("line3.pdf", dpi = 1000)
plt.savefig("line4.jpg")
plt.savefig("line5",transparent = True,facecolor = "b")
plt.savefig("line6",bbox_inches = "tight",facecolor ="g")

plt.show()
```
