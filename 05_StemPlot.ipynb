# Stem Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [3,4,5,2,1,3]
plt.stem(x,y,linefmt = ":",markerfmt = "r+", bottom = 3, basefmt = "g", label = "Lakshay",orientation = "vertical")
plt.legend()
plt.show()
```
