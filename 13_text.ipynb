# text

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,19,15,13,12]
plt.plot(x,y)
plt.title("Lakshay",fontsize = 20)
plt.xlabel("Days",fontsize = 15)
plt.ylabel("hours",fontsize = 15)

plt.text(2,13,"java",fontsize = 15, style = "italic", bbox = {"facecolor":"red"})
plt.annotate("Lucky", xy=(1.5,12), xytext = (4,14),arrowprops= {"facecolor" :"black","shrink" : 100})
plt.legend(["Ggn"], loc = 0, facecolor = "r",edgecolor = "g",framealpha = 0.7,shadow = True)#0-10
plt.show()
```
